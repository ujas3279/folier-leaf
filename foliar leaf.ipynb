```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
DataDir = "H:/sgp/Foliar leaf Diesease/images/"
Class_Name = ["healthy","multiple_diseases","rust","scab"]

training_data = []
Traning_Images = []
Traning_Labels = []
```

```python
def cerate_training_data():
    count = 0
    for clas in Class_Name:
        path = os.path.join(DataDir,clas)
        class_num = Class_Name.index(clas)
        for img in os.listdir(path):
            img_array = cv2.imread(os.path.join(path,img))#, cv2.IMREAD_GRAYSCALE
            img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
            new_array = cv2.resize(img_array, (1024,683))
            data = np.asarray(new_array)
            print(img)
            print("Train image: ",count)
            training_data.append([data, class_num])
            Traning_Images.append(data)
            Traning_Labels.append(np.array(class_num))
            count = count + 1
```

```python
cerate_training_data()
```

```python
np.save("Traning_Images",np.array(Traning_Images))
np.save("Traning_Labels",np.array(Traning_Labels))
```
